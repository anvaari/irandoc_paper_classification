```python
#Import PAckages
import numpy as np
import pandas as pd
from numpy import NaN as nan
import hazm
```

```python
def make_text_clean(text):
    return text.replace('\xad','').replace('\u200c','').replace('\u200f','').replace('\n',' ').replace('\xa0','').replace('\u200e','').replace('\r','').replace('\u202b','') 
```

```python
#Loda Data
data=pd.read_excel('data.xlsx')
data.rename(columns={'Unnamed: 0':'ID' },inplace=True)

#drop nan values
data.dropna(subset=['Abstract'],inplace=True)
data.head(10)
```

```python
#See Unique subjects
unique_subject=data['Subject'].value_counts()
unique_subject
```

```python
#asign subject with lower instance with "سایر"

#make dict of them
new_subject=dict()
for name in unique_subject.index:
    clean_name=make_text_clean(name)
    if unique_subject[name]<10:
        new_subject[name]='سایر'
    else:
        new_subject[name]=name
#replace in dataframe
data.replace(to_replace={'Subject':new_subject},value=None,inplace=True)
unique_subject=data['Subject'].value_counts()
unique_subject
```

```python
#find most frequesnt words in each subject
stop_words= hazm.utils.stopwords_list(stopwords_file='/home/anvaari/Python Packages/hazm/hazm/data/stopwords.dat')
signs=['،','«','»','.',')','(','"',':',';','%','-','?',',','؛']
tokenizer = hazm.WordTokenizer()
tokenizer_sentenc=hazm.SentenceTokenizer()
ketab=pd.DataFrame(data[data['Subject']==unique_subject.index[0]])
ketab.dropna(subset=['Subject'],inplace=True)
fanav=pd.DataFrame(data[data['Subject']==unique_subject.index[1]])
fanav.dropna(subset=['Subject'],inplace=True)
other=pd.DataFrame(data[data['Subject']==unique_subject.index[2]])
other.dropna(subset=['Subject'],inplace=True)
```

```python
#First subject
frequent_word_ketab=dict()
absts_sentenc=[]
for abst in ketab['Abstract']:
    absts_sentenc.append(tokenizer_sentenc.tokenize(abst))
    
for abst in absts_sentenc:
    for sentence in abst:
        sentence=make_text_clean(sentence)
        for word in tokenizer.tokenize(sentence):
            if word not in signs and word not in stop_words and word not in frequent_word_ketab:
                frequent_word_ketab[word]=1
            elif word in frequent_word_ketab:
                frequent_word_ketab[word]+=1
sorted(frequent_word_ketab.items(), key=lambda x:x[1],reverse=True)[0:10]
```

```python
#Describe of Frequency of 'فناوری' in first subject
fanavari_freq_ketab=dict()
for index,row in ketab.iterrows():
    abst=make_text_clean(row['Abstract'])
    words_count=0
    fanavari_count=0
    for word in hazm.word_tokenize(abst):
        if word not in signs and word not in stop_words:
            words_count+=1
        if word == 'اطلاعات':
            fanavari_count+=1
    fanavari_freq_ketab[index]=fanavari_count
pd.Series(fanavari_freq_ketab).describe()
```
